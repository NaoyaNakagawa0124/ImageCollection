```python
import requests
import bs4
import os
from pykakasi import kakasi

# pykakasiの初期設定
kakasi = kakasi()
kakasi.setMode('H', 'a')  # ひらがな・カタカナをローマ字に
conv = kakasi.getConverter()

# 検索URL（Sakurazaka46）
search_url = 'https://sakurazaka46.com/s/s46/search/artist?ima=0000'
response = requests.get(search_url)
response.raise_for_status()

# BeautifulSoupでHTMLパース
bs4_blog = bs4.BeautifulSoup(response.text, 'html.parser')

# メンバーリストの取得
all_member = bs4_blog.select('.sort-default li')

member_No = []
name_kanji = []
name_rome = []

# データフォルダの作成（存在しない場合のみ作成）
if not os.path.exists('data'):
    os.mkdir('data')

# メンバーごとの処理
for i in range(len(all_member)):
    # メンバー番号の取得
    member_No.append(all_member[i]['data-member'])

    # メンバーの漢字名の取得（class="name"）
    name_kanji.append(all_member[i].find('p', {'class': 'name'}).get_text().strip())

    # メンバーごとのディレクトリの作成（存在しない場合のみ作成）
    path = os.path.join('data', name_kanji[i])
    if not os.path.exists(path):  # ディレクトリが存在しない場合のみ作成
        os.mkdir(path)
    else:
        print(f"{path} ディレクトリはすでに存在しています")

    # メンバーのカナ名の取得（class="kana"）
    name_kana = all_member[i].find('p', {'class': 'kana'}).get_text().strip()
    print(name_kana)

    # 名前のローマ字変換
    name_parts = name_kana.split()
    if len(name_parts) == 2:
        last = conv.do(name_parts[0])
        first = conv.do(name_parts[1])
    else:
        last = conv.do(name_kana)  # 名前が一部分しかない場合

    # ローマ字名のリストに追加
    name_rome.append(first + '.' + last)

# ローマ字リストの確認
print(name_rome)
```

```python
import os
import requests
from bs4 import BeautifulSoup
from pykakasi import kakasi
from PIL import Image
from io import BytesIO

# pykakasiの新しいAPIの設定
kks = kakasi()  # kakasiのインスタンス作成
conv = kks.getConverter()  # コンバータ取得

try:
    url = 'https://sakurazaka46.com/s/s46/diary/detail/57068?ima=0000&cd=blog'

    # ヘッダーを追加してリクエストを送信
    headers = {'User-Agent': 'Mozilla/5.0'}
    response = requests.get(url, headers=headers)
    response.raise_for_status()

    print(f"ステータスコード: {response.status_code}")

    bs4_blog = BeautifulSoup(response.text, 'html.parser')

    # 記事のセレクタを修正
    articles = bs4_blog.select('article.post')

    print(f"取得した記事の数: {len(articles)}")

    # 保存ディレクトリを作成
    save_dir = os.path.join('data', 'テスト')
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)

    # 記事ごとに処理
    for article in articles:
        # メンバー名の取得
        member_name_element = article.find('p', {'class': 'name'})
        if member_name_element:
            member_name_kanji = member_name_element.get_text().strip()
            print(f"メンバー名（漢字）: {member_name_kanji}")
            # ローマ字に変換
            member_name_rome = conv.do(member_name_kanji)  # ローマ字に変換
            print(f"メンバー名（ローマ字）: {member_name_rome}")
        else:
            print("メンバー名が見つかりませんでした")
            member_name_rome = "unknown_member"

        # 年、月、日の取得
        year_element = article.find('span', {'class': 'ym-year'})
        month_element = article.find('span', {'class': 'ym-month'})
        day_element = article.find('p', {'class': 'date wf-a'})

        if year_element and month_element and day_element:
            year = year_element.get_text().strip()
            month = month_element.get_text().strip().zfill(2)  # 月を2桁にする
            day = day_element.get_text().strip().zfill(2)  # 日を2桁にする
            date = f"{year}/{month}/{day}"
            print(f"日付: {date}")
        else:
            print("日付が見つかりませんでした")
            date = "unknown_date"

        # 画像の取得
        img_elements = article.select('div.box-article img')
        if img_elements:
            img_counter = 1  # 画像ごとに連番を付けるためのカウンタ
            for img in img_elements:
                src = img.get('src')
                print(f"画像URL: {src}")

                # フルURLに変換
                img_url = f'https://sakurazaka46.com{src}'
                # 画像ファイル名にメンバー名、日付、連番を使用
                img_name = f"{member_name_rome}_{date.replace('/', '_')}_{img_counter}.jpeg"
                img_path = os.path.join(save_dir, img_name)

                # 画像をダウンロード
                img_response = requests.get(img_url)
                if img_response.status_code == 200:
                    # 画像をPillowで開く
                    img_data = Image.open(BytesIO(img_response.content))

                    # 解像度を変更（例: 元の2倍に拡大）
                    new_size = (img_data.width * 2, img_data.height * 2)
                    img_resized = img_data.resize(new_size, Image.Resampling.LANCZOS)  # Pillow 10.0以降はResampling.LANCZOSを使用

                    # 画像を保存
                    img_resized.save(img_path)
                    print(f"画像を保存: {img_path}")
                else:
                    print(f"画像のダウンロードに失敗: {img_url}")

                # 連番をインクリメント
                img_counter += 1
        else:
            print("画像が見つかりませんでした")

except Exception as e:
    print(f"エラーが発生しました: {e}")
```

```python
import os
import requests
from bs4 import BeautifulSoup
from pykakasi import kakasi
from PIL import Image
from io import BytesIO
import time
import random

# pykakasiの設定
kks = kakasi()
conv = kks.getConverter()

# メンバー名とそのブログトップページのURLを取得する関数
def get_member_list(base_url):
    headers = {'User-Agent': 'Mozilla/5.0'}
    response = requests.get(base_url, headers=headers)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, 'html.parser')
    members = soup.select('ul.com-blog-circle li a')

    member_links = []
    for member in members:
        member_name = member.select_one('p.name').get_text().strip()
        member_url = f"https://sakurazaka46.com{member['href']}"
        member_links.append((member_name, member_url))

    return member_links

# ブログ記事のURLを取得する関数（ページネーション対応）
def get_blog_list(page_url, member_name):
    headers = {'User-Agent': 'Mozilla/5.0'}
    response = requests.get(page_url, headers=headers)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, 'html.parser')

    blog_links = []
    articles = soup.select('ul.com-blog-part li.box a')
    for article in articles:
        blog_link = f"https://sakurazaka46.com{article['href']}"
        blog_links.append(blog_link)

    # 次のページがあれば取得
    next_page_element = soup.select_one('li > a[rel="next"]')  # 「→」ボタン
    next_page_url = None
    if next_page_element:
        next_page_url = f"https://sakurazaka46.com{next_page_element['href']}"

    return blog_links, next_page_url

# 各ブログページをスクレイピングして画像を保存する関数
def scrape_blog_page(blog_url, member_name_rome, save_dir):
    try:
        headers = {'User-Agent': 'Mozilla/5.0'}
        response = requests.get(blog_url, headers=headers)
        response.raise_for_status()

        bs4_blog = BeautifulSoup(response.text, 'html.parser')
        articles = bs4_blog.select('article.post')

        for article in articles:
            # 日付の取得
            year_element = article.find('span', {'class': 'ym-year'})
            month_element = article.find('span', {'class': 'ym-month'})
            day_element = article.find('p', {'class': 'date wf-a'})

            if year_element and month_element and day_element:
                year = year_element.get_text().strip()
                month = month_element.get_text().strip().zfill(2)
                day = day_element.get_text().strip().zfill(2)
                date = f"{year}/{month}/{day}"
            else:
                date = "unknown_date"

            # 画像の取得
            img_elements = article.select('div.box-article img')
            if img_elements:
                img_counter = 1
                for img in img_elements:
                    src = img.get('src')
                    img_url = f'https://sakurazaka46.com{src}'
                    img_name = f"{member_name_rome}_{date.replace('/', '_')}_{img_counter}.jpeg"
                    img_path = os.path.join(save_dir, img_name)

                    img_response = requests.get(img_url)
                    if img_response.status_code == 200:
                        img_data = Image.open(BytesIO(img_response.content))
                        new_size = (img_data.width * 2, img_data.height * 2)
                        img_resized = img_data.resize(new_size, Image.Resampling.LANCZOS)
                        img_resized.save(img_path)
                        print(f"画像を保存: {img_path}")
                    else:
                        print(f"画像のダウンロードに失敗: {img_url}")

                    img_counter += 1
                    time.sleep(random.uniform(2, 5))

    except Exception as e:
        print(f"エラーが発生しました: {e}")

# メンバーごとの全ブログをスクレイピング
def scrape_all_blogs(member_url, member_name_rome, member_name_kanji):
    next_page_url = member_url
    save_dir = os.path.join('data', member_name_kanji)
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)

    while next_page_url:
        print(f"ページをスクレイピング中: {next_page_url}")
        blog_links, next_page_url = get_blog_list(next_page_url, member_name_kanji)
        for blog_url in blog_links:
            scrape_blog_page(blog_url, member_name_rome, save_dir)
            time.sleep(random.uniform(2, 5))

# メイン処理
if __name__ == "__main__":
    base_url = 'https://sakurazaka46.com/s/s46/diary/blog/list?ima=0000'
    member_list = get_member_list(base_url)

    for member_name, member_url in member_list:
        member_name_rome = conv.do(member_name)  # ローマ字に変換
        scrape_all_blogs(member_url, member_name_rome, member_name)
```
